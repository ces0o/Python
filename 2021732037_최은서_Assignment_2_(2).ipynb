```python
'''
코스피 종가 데이터를 자동으로 가져와 그래프로 그려주는 스크립트 입니다.
pandas_datareader, yfinance, matplotlib 라이브러리를 설치해줘야 합니다.
* 설치방법 (터미널 기준)
pip install pandas_datareder
pip install yfinance
pip install matplotlib
'''

from pandas_datareader import data as pdr # 라이브러리에서 Data 모듈을 pdr로 가져온다
from matplotlib import pyplot as plt # 라이브러리에서 pyplot 모듈을 plt로 가져온다
import yfinance as yf # 라이블리를 yf로 가져온다
yf.pdr_override() # 라이브러리 설정 변경

df = pdr.get_data_yahoo("^KS11") # 라이브러리를 사용하여 KS11에 대한 주식 데이터를 가져와서 df에 저장함
y_values = df['Adj Close'] # DataFrame에서 Adj Close 열을 추출하여 주식 종가 데이터를 변수에 저장
x_values = range(len(y_values)) # 길이에 따라 일련의 정수 값을 생성하여 x축 데이터를 변수에 저장

plt.plot(x_values,y_values) # matplotlib을 사용하여 x축, y축에 선 그래프를 그린다
plt.show() # 그래프를 화면에 표시
```

```python
'''
다음 헤드라인 뉴스를 자동으로 가져오는 스크립트 입니다.
별도 라이브러리를 설치하지 않아도 실행 가능합니다.
'''
import urllib.request # Python 내장 모듈인 urllib.request를 가져온다
import html.parser # HTML 문자열을 파싱하기 위한 html.parser 모듈을 가져온다
u = 'http://www.daum.net' # 변수 u에 웹 페이지 URL인 "http://www.daum.net"을 할당
f = urllib.request.urlopen(u) # urllib.request.urlopen() 함수를 사용하여 지정된 URL에서 웹 페이지를 연다
contents = str(f.read().decode('utf-8'))
# f.read()를 사용하여 웹 페이지의 내용을 바이트로 읽어온 다음, decode('utf-8')을 사용하여 UTF-8 인코딩으로 디코딩하고 문자열로 변환하여 contents 변수에 저장
f.close()
# 웹 페이지를 열었던 파일 핸들을 닫는다
i,k = 0,0
# 변수 i와 k를 0으로 초기화

# infinite loop
while True:
  # 무한 루프를 시작
    k = k + 1
    # 루프 반복 횟수를 기록하기 위해 변수 k를 1 증가
    i = contents.find('뉴스1탭_텍스트소재', i+1)
    #  문자열 contents에서 '뉴스1탭_텍스트소재'를 찾아서 그 위치를 변수 i에 저장
    if i == -1 or k > 20:
      # 만약 '뉴스1탭_텍스트소재'를 더 이상 찾지 못하거나 k가 20을 초과하면 다음 단계로 이동
        break
        # 무한 루프를 종료
    i = contents.find('">', i+1)
    # 뉴스1탭_텍스트소재' 다음에 나오는 ">" 문자열을 찾아서 그 위치를 변수 i에 저장
    j = contents.find('</a>', i+1)
    # 다음에 나오는 "</a>" 문자열을 찾아서 그 위치를 변수 j에 저장
    data = contents[i+2:j]
    # i와 j 사이의 문자열을 추출하여 data 변수에 저장
    print(html.unescape(data))
    # data 변수에 저장된 HTML 문자열을 파싱하고 특수 문자를 해석하여 출력
```

```python
for i in range(5):
    if i==3:
      # i가 3인 경우 아래의 코드 블록을 실행
        break
        # 조건이 충족되면 바깥쪽 for 루프를 종료하고, 루프를 벗어난다
    for j in range(5):
      # j 변수는 0부터 4까지의 값을 가진다
        if j==2:
          # 안쪽 for 루프에서 j의 값을 확인하는 조건문
            break
            # 조건이 충족되면 안쪽 for 루프를 종료하고, 루프를 벗어난다
        print(i, j)
        #  안쪽 for 루프에서 i와 j의 값을 출력
```

```python
i = [1,2,3,4,-1,5,6,7,8]
j = []
count=0
# 변수를 0으로 초기화
while(i[count]>0):
# while 루프를 시작하고, 루프 조건은 i[count]가 양수인 동안 반복
    j.append(i[count])
    # i[count] 값을 j 리스트에 추가
    count = count + 1
print(j)
```
