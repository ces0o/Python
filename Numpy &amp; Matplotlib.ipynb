###  실습 (Assignment 3)
* 행렬 방정식 풀기
  - 다음 행렬 방정식을 'Pseudo inverse matrix'를 이용해 풀어보자
  - $A^{T}A$의 역행렬이 존재한다고 가정
$$Ax=B$$
$$A = \begin{bmatrix}0 & 1 \\ 1 & 1 \\ 2 & 1 \\ 3 & 1 \end{bmatrix} $$  
$$B = \begin{bmatrix}-1 \\ 0.2 \\ 0.9 \\ 2.1 \end{bmatrix} $$


```python
import torch

A = torch.FloatTensor([[0,1],[1,1],[2,1],[3,1]]) # A 행렬 선언
b = torch.FloatTensor([-1,0.2,0.9,2.1]) # b 행렬 선언
Tb = b.reshape(4,1) # 계산해주기 위해 b 행렬을 4,1로 reshape한 것을 Tb로 선언
AA = (A.T).matmul(A) # trans한 A와 A를 곱한 것을 AA로 선언
TAA = torch.inverse(AA) # AA의 역행렬을 구하고 그것을 TAA로 선언
AT = A.T # A의 trans를 AT로 선언
ATb = AT.matmul(Tb) # AT와 Tb의 곱을 ATb로 선언
print(TAA.matmul(ATb)) # TAA와 ATb의 곱을 프린트함으로써 결과값 출력
```

###  실습 (Assignment4)
* 2x2 그래프 그리기
  - $y=x$
  - $y=x^2$
  - $y=sin(x)$
  - $y=cos(x)$


```python
import numpy as np
import matplotlib.pyplot as plt
# 그래프 그려주는 모듈 선언

x = np.linspace(-2*np.pi, 2*np.pi, 100)
# -2π부터 2π까지 범위에서 총 100개의 숫자를 생성
y1 = x
y2 = x**2
y3 = np.sin(x)
y4 = np.cos(x)
# y값 계산

# 2x2 서브플롯을 생성
plt.figure(figsize=(12, 8))
#가로 너비 : 12 인치 / 세로 높이: 8 인치

# 첫 번째 서브플롯 (y = x)
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('y = x')

# 두 번째 서브플롯 (y = x^2)
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('y = x^2')

# 세 번째 서브플롯 (y = sin(x))
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('y = sin(x)')

# 네 번째 서브플롯 (y = cos(x))
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('y = cos(x)')

# 그래프를 출력
plt.tight_layout()
plt.show()
```
